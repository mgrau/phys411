# Grade Calculator
I will calculate the course GPA from the final course grade according to the following formula:
$$
\text{GPA} = \frac{\text{Grade} - 60\%}{93\%-60\%} \times (4.0 - 0.7) + 0.7
$$

```python
from ipywidgets import interact
```

```python
def GPA(grade):
    return (grade-0.6)/(0.93-0.6) * (4-0.7) + 0.7

def letter_grade(GPA):
    if GPA >= 4.0:
        return 'A'
    elif GPA >= 3.7:
        return 'A-'
    elif GPA >= 3.3:
        return 'B+'
    elif GPA >= '3.0':
        return 'B'
    elif GPA >= '2.7':
        return 'B-'
    elif GPA >= '2.3':
        return 'C+'
    elif GPA >= '2.0':
        return 'C'
    elif GPA >= '1.7':
        return 'C-'
    elif GPA >= '1.3':
        return 'D+'
    elif GPA >= '1.0':
        return 'D'
    elif GPA >= '0.7':
        return 'D-'
    else:
        return 'F'
```
